```python
grade = int(input("Hello there, please input your finaL average, and i will tell you your final letter outcome"))

if grade >= 90 and grade <= 100:
  print ("Congratulations, you got an A")

elif grade >= 80 and grade <= 89:
  print ("Nice, yougot a B")

elif grade >= 70 and grade <= 79:
  print("Not bad, you got a C")

elif grade >= 0 and grade <= 69:
  print("oh no, you got an F")

else:
  print("invalid input")
```
