```python
correct_password = "secure555"
max_attempts = 3
attempts = 0


while attempts < max_attempts:
  user_input = input("Enter the password: ")
  attempts += 1

  if user_input == correct_password:
    print("Access granted!")
    break
  else:
    print("Incorrect password. Attempts left:", max_attempts - attempts)
```
